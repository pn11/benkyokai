# Octave 入門

Jupyter で動かすには `OCTAVE_EXECUTABLE` の環境変数を設定する。

- [octave_kernel · PyPI](https://pypi.org/project/octave_kernel/)

```octave
# 画像の読み込み

img=imread("lena.bmp");
```

```octave
# カラー画像が表示されない
imshow(img)
```

```octave
# ダブル画像にすると表示できる cf. https://github.com/Calysto/octave_kernel/issues/86
imshow(im2double(img))
```

```octave
# 関数、プリント、戻り値の受け取り方
function print_size(img)
    [h,w,d]=size(img);
    printf("%dx%dx%d\n", h, w, d);
endfunction
```

```octave
print_size(img);
```

```octave
# グレースケール変換
gray = uint8(mean(img,3));
imshow(gray)
```

```octave
# Reshape
array = reshape(img, 1, 512*512*3);
```

```octave
# Histogram
hist(array, 256)
```

```octave
# Edge detection
pkg load image
edge(gray, "Canny") # https://octave.sourceforge.io/image/function/edge.html
```

```octave
edge(gray, "LoG")
```
